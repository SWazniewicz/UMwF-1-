2.2. W rezultacie wyniki testów mogą znacząco się od siebie różnić.

```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error

R_train = R.copy()

user_means = R_train.replace(0, np.nan).mean(axis=1)

R_bias = R_train.sub(user_means, axis=0).fillna(0)

np.random.seed(42)
test_indices = []
for user in R.index:
    rated_movies = R.loc[user][R.loc[user] > 0].index
    if len(rated_movies) > 0:
        movie = np.random.choice(rated_movies)
        R_train.loc[user, movie] = 0
        test_indices.append((user, movie))
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(R_bias)
np.fill_diagonal(similarity, 0)
similarity_df = pd.DataFrame(similarity, index=R_train.index, columns=R_train.index)

def predict_knn_means(user_id, item_id, k=5):
    user_mean = R_train.loc[user_id][R_train.loc[user_id] > 0].mean()

    users_who_rated = R_train[R_train[item_id] > 0].index
    if len(users_who_rated) == 0:
        return user_mean

    sim_scores = similarity_df.loc[user_id, users_who_rated]
    top_k_users = sim_scores.nlargest(k).index

    preds = []
    for neighbor in top_k_users:
        neighbor_mean = R_train.loc[neighbor][R_train.loc[neighbor] > 0].mean()
        preds.append(R_train.loc[neighbor, item_id] - neighbor_mean + user_mean)

    return np.mean(preds)
```

```python
import math

results = []

for k in range(3, 11):
    y_true = []
    y_pred = []
    for user, movie in test_indices:
        y_true.append(R.loc[user, movie])
        y_pred.append(predict_knn_means(user, movie, k=k))

    rmse = math.sqrt(mean_squared_error(y_true, y_pred))
    results.append((k, rmse))

results_df = pd.DataFrame(results, columns=['k', 'RMSE'])
print(results_df)
best_k = results_df.loc[results_df['RMSE'].idxmin(), 'k']
print("Najlepsza liczba sąsiadów:", best_k)
```

2.3. RMSE maleje wraz z wzrostem liczby sąsiadów i jest najmniejszy dla 10, dlatego powinno się użyć w modelu 10 najbliższych sąsiadów.

```python
import numpy as np
from sklearn.metrics import mean_squared_error
import math
```

```python
def cross_validate_knn_means(R, k_neighbors=5, k_folds=5, random_state=42):
    np.random.seed(random_state)
    all_ratings = [(u, m) for u in R.index for m in R.columns if R.loc[u, m] > 0]
    np.random.shuffle(all_ratings)

    fold_size = len(all_ratings) // k_folds
    rmses = []

    for i in range(k_folds):
        test_fold = all_ratings[i*fold_size:(i+1)*fold_size]
        train_fold = [r for r in all_ratings if r not in test_fold]

        R_train = R.copy()
        for u, m in test_fold:
            R_train.loc[u, m] = 0

        user_means = R_train.replace(0, np.nan).mean(axis=1)
        R_bias = R_train.sub(user_means, axis=0).fillna(0)
        similarity = cosine_similarity(R_bias)
        np.fill_diagonal(similarity, 0)
        similarity_df = pd.DataFrame(similarity, index=R_train.index, columns=R_train.index)

        def predict_fold(user_id, item_id):
            user_mean = R_train.loc[user_id][R_train.loc[user_id] > 0].mean()
            users_who_rated = R_train[R_train[item_id] > 0].index
            if len(users_who_rated) == 0:
                return user_mean
            sim_scores = similarity_df.loc[user_id, users_who_rated]
            top_k_users = sim_scores.nlargest(k_neighbors).index
            preds = []
            for neighbor in top_k_users:
                neighbor_mean = R_train.loc[neighbor][R_train.loc[neighbor] > 0].mean()
                preds.append(R_train.loc[neighbor, item_id] - neighbor_mean + user_mean)
            return np.mean(preds)

        y_true = [R.loc[u, m] for u, m in test_fold]
        y_pred = [predict_fold(u, m) for u, m in test_fold]
        rmse = math.sqrt(mean_squared_error(y_true, y_pred))
        rmses.append(rmse)
        print(f"Fold {i+1}: RMSE = {rmse:.4f}")

    mean_rmse = np.mean(rmses)
    std_rmse = np.std(rmses)
    print(f"\nŚrednie RMSE: {mean_rmse:.4f}, Odchylenie standardowe: {std_rmse:.4f}")
    return mean_rmse, std_rmse
```

```python
mean_rmse, std_rmse = cross_validate_knn_means(R, k_neighbors=10, k_folds=5)
```

2.4. Średnio przewidywana ocena odbiega od rzeczywistej wartości o prawie 1 w skali ocen 1-5. Bardzo małe odchylenie standardowe wskazuje, że wynik RMSE jest bardzo stabilny w różnych podziałach danych.
